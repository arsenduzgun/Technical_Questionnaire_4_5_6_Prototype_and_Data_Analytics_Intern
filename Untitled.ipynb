```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("country_vaccination_stats.csv")
```

```python
#Q4
dailyVacDict = dict()
for i in df.groupby("country").min().iterrows(): dailyVacDict[i[0]] = 0 if np.isnan(i[1][1]) else i[1][1]
df["daily_vaccinations"] = df.apply(lambda row: dailyVacDict[row.country] if np.isnan(row.daily_vaccinations) else row.daily_vaccinations, axis = 1)
```

```python
#Q5
df.groupby("country").median().sort_values("daily_vaccinations", ascending = False)[:3]
```
